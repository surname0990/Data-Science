2.  Создайте класс сотрудника Employee. При инициализации класса задается имя сотрудника name и его текущая зарплата salary. Напишите следующие методы:
Метод up, который увеличивает зарплату сотрудника на 100
Метод print, который выводит на экран текущую зарплату сотрудника в формате "Сотрудник Иван, зарплата 100"

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary


    def print_(self):
        print('Сотрудник {}, получает {}'.format(self.name,self.salary))

    def up(self):
        self.salary = self.salary + 100
        self.print_()

Employee("kolya", 0).up()
Employee("vasya", 100).print_()

```

```python
#перечень полок
directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
def s():           #Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится
    check = True 
    doc = input()  
    for item in directories.keys(): #находим ключи (1,2,3)
        for value in directories.get(item): #находим значение подставив ключ 
            if value == doc:
                print("Документ хранится на полке: ", item)
                check = False
    if check:
        print("Документ не найден в базе")

s()
```

Пункт 4. Пользователь по команде “ads” может добавить новую полку
Примеры работы:

Введите команду:
ads

Введите номер полки:
10
Результат:
Полка добавлена. Текущий перечень полок: 1, 2, 3, 10.

Введите команду:
ads

Введите номер полки:
1
Результат:
Такая полка уже существует. Текущий перечень полок: 1, 2, 3.

```python
def ads(dock):
    print('Результат:')
    check = True
    for value in directories.keys():
        if value == dock:
            print('Такая полка уже существует. Текущий перечень полок:', end=" ")
            printitem()
            check = False
    directories.update({dock: []})
    if check:
        print('Полка добавлена. Текущий перечень полок:', end=" ")
        printitem()


def printitem():
    print(*[item for item in directories.keys()], sep=', ', end=".\n")

ads(10)
```

Пункт 5. Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая)
Примеры работы:

Введите команду:
ds

Введите номер полки:
3
Результат:
Полка удалена. Текущий перечень полок: 1, 2.

Введите команду:
ds

Введите номер полки:
1
Результат:
На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: 1, 2, 3.

Введите команду:
ds

Введите номер полки:
4
Результат:
Такой полки не существует. Текущий перечень полок: 1, 2, 3.

```python
def ds(dock):
    if directories.get(dock) == []:
        directories.pop(dock)
        print('Полка удалена. Текущий перечень полок:', end=" ")
        printitem()
    elif directories.get(dock) == None:
        print('Такой полки не существует. Текущий перечень полок:', end=" ")
        printitem()
    else:
        print('На полке есть документа, удалите их перед удалением полки.', end=" ")
        printitem()
ds(10)
```
